```python
import random
def fitness_goals():
  goals = ["lose weight", "gain weight", "tone up"]
  name = input("What is your name?: ")
  print()

  userinput = input("What is your fitness goal?: ").lower()
  print()

  if userinput in goals:
    conf = input(f'{name} wants to {userinput.lower()}?: ').lower()
    if conf == "yes":
      print()
      print("Great! We can help you get started!")
      return userinput
    elif conf == "no":
       print("Mind redefining your goals?")
       return None
  else:
       print("Goal not found. Please pick one:")
       for g in goals:
        print("-", g)
       return None




def exercise_categories():
  category = ["Aerobics", "Strength", "Cardio", "Flexibility", "Balance"]
  print()
  user = input("What type of exercise would you like to do?: ").title()

  if user in category:
     print()
     print(f"We can help you develop a plan to improve {user.lower()}!")
     print()
     return user
  else:
     print(f"Category not found. Please pick one:")
     for c in category:
      print("-", c)
      print()
  return None


def workout_plan(goal, category):
   print()
   if not goal or not category:
        print("Cannot create a workout plan without valid goal and category.")
        print()
        return
   aer_exercises = ["Elliptical Machine", "Stationary Biking", "Rowing Machine", "Elliptical Trainer"]
   cardio = ["Treadmill", "Cycling", "HIIT", "Swimming"]
   strength = ["Squats", "Bench Press", "Deadlifts", "Cable Rows"]
   flexibility = ["Pilates", "Yoga Flow", "Core Circuits", "Hip Flexor Stretch"]
   balance = ["Single Leg Stands", "Heel-to-Toe Walking", "Standing Leg Lifts", "Bird Dog"]

   exercises = {
    "Aerobics": aer_exercises,
    "Cardio": cardio,
    "Strength": strength,
    "Flexibility": flexibility,
    "Balance": balance }

   if category in exercises:
     print(f"\n--- Workout Plan for: {goal} using {category} training ---\n")
     print("Our suggested exercises are:")
     for ex in random.sample(exercises[category], 3):
        print("-", ex)



goal_choice = fitness_goals()
category = exercise_categories()
workout_plan(goal=goal_choice, category=category)
```
