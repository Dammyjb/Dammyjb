```python
def fitness_goals():
  goals = ["Lose weight", "Gain weight", "Tone up"]
  name = input("What is your name?: ")

  userinput = input("What is your fitness goal?: ")

  if userinput in goals:
    conf = input(f'{name} wants to {userinput.lower()}?: ').lower()
    if conf == "yes":
       print("Great! We can help you get started!")
       return userinput
    elif conf == "no":
       print("Mind redefining your goals?")
       return None
  else:
       print("Goal not found. Please pick one:")
       for g in goals:
        print("-", g)
       return None




def exercise_categories():
  category = ["Aerobics", "Strength", "Cardio", "Flexibility", "Balance"]

  user = input("What type of exercise would you like to do?: ")

  if user in category:
     print(f"We can help you develop a plan to improve {user.lower()}!")
     return user
  else:
     print(f"Category not found. Please pick one:")
     for c in category:
      print("-", c)
     input()
  return None


def workout_plan(goal, category):
   if not goal or not category:
        print("Cannot create a workout plan without valid goal and category.")
        return
   print(f"\n--- Workout Plan for: {goal} using {category} training ---")
   if goal == "Lose weight" and category == "Cardio":
      print("• 30 mins treadmill\n• 20 mins cycling\n• 10 mins HIIT")
   elif goal == "Lose weight" and category == "Strength":
      print("• Squats\n• Bench press\n• Deadlifts\n• 3–5 reps, heavy weight")
   elif goal == "Tone up" and category == "Flexibility":
      print("• Pilates\n• Yoga flow\n• Core circuits")
   else:
      print("• Mix of exercises based on your preferences.")



goal_choice = fitness_goals()
category = exercise_categories()
workout_plan(goal=goal_choice, category=category)
```
